```python
import numpy as np
from classifiers import FGMBinaryClassifier
from sklearn.model_selection import train_test_split
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
X = np.random.randn(1000, 4)
y = np.sign(X[:, 0]**2 - X[:, 1]**-2 + np.sin(X[:, 2]) - 10*X[:, 3])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

### Logistic Regression

```python
lmbda = 0.5
clf = FGMBinaryClassifier(classifier='logistic', lmbda=lmbda, max_iter=10)
clf.fit(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

```python
train_obj = [clf._classifier.computeobj(X_train, y_train, beta, lmbda) for beta in clf._betas]
test_obj = [clf._classifier.computeobj(X_test, y_test, beta, lmbda) for beta in clf._betas]
```

```python
plt.figure(figsize=(16, 6))
plt.plot(train_obj, marker='o')
plt.plot(test_obj, marker='o')
plt.xlabel("Iterations")
plt.ylabel("Objective Function")
plt.grid()
plt.title("Objective Function vs Iterations")
```

### Linear SVM

```python
lmbda = 0.5
clf = FGMBinaryClassifier(classifier='svm', lmbda=lmbda, max_iter=10)
clf.fit(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

```python
train_obj = [clf._classifier.computeobj(X_train, y_train, beta, lmbda) for beta in clf._betas]
test_obj = [clf._classifier.computeobj(X_test, y_test, beta, lmbda) for beta in clf._betas]
```

```python
plt.figure(figsize=(16, 6))
plt.plot(train_obj, marker='o')
plt.plot(test_obj, marker='o')
plt.xlabel("Iterations")
plt.ylabel("Objective Function")
plt.grid()
plt.title("Objective Function vs Iterations")
```
